```python
start = pd.to_datetime('2023-01-01')
eval_dates = [start + datetime.timedelta(days=7*j) for j in range(0,52)]

cases = pd.read_csv('/Users/elizabethmurphy/Downloads/covid19cases_test.csv',index_col=0)
cases = cases[cases['area'] =='San Diego']
cases = cases[(cases.index>='2020-06-01') & (cases.index<='2022-07-01')]
cases.index = pd.to_datetime(cases.index)

cases['cases'] = cases['cases'].rolling(window=7, center=True, min_periods=0).mean()
cases['positivity'] = cases['cases'].rolling(window=7, center=True, min_periods=0).mean()/cases['total_tests'].rolling(window=7, center=True, min_periods=0).mean()

ww_dict = {'Point Loma':['/Users/elizabethmurphy/Downloads/PointLoma_sewage_seqs.csv','/Users/elizabethmurphy/Downloads/PointLoma_sewage_qPCR.csv']}
for site, files in zip(ww_dict.keys(),ww_dict.values()):
    df = pd.read_csv(f'{files[0]}')

    df['Date'] = pd.to_datetime(df['Date'])
    df.columns = [dfc.split(' (')[0] for dfc in df.columns]
    df =df.set_index('Date')
    df = df[df.index>='2020-06-01']
    df = df[df.index<='2022-07-01']
    df = df.dropna(axis = 0, how = 'all')
    df = df.fillna(0)
    df = df/100.

    df = df.drop(columns=['Other'])
    df = df[df.columns[df.sum(axis=0) > 0.01]]

    cdf = pd.read_csv(f'{files[1]}')
    cdf['Sample_Date'] = pd.to_datetime(cdf['Sample_Date'])
    cdf =cdf.set_index('Sample_Date')
    sharedInds = np.sort(list(set(cdf.index) & set(df.index)))
    cdf = cdf.loc[sharedInds]
    df = df.loc[sharedInds]
    df = df[~df.index.duplicated(keep='last')]
    scaleddf = df.mul(cdf['Mean viral gene copies/L'],axis=0)

    cdf = cdf.resample('D').asfreq()
    cdf = cdf.rolling(window=7, center=True, min_periods=0).mean()


cdf = pd.concat([cdf,cases],axis=1)
cdf = cdf.dropna(how='any')

N = 30
F = 5
# first, let's learn the shedding kernel
X = np.array([cdf['Mean viral gene copies/L'].values[(j-F):(j+N-F)] for j in range(F,(cdf.shape[0]-N+F))])/cdf['Mean viral gene copies/L'].mean()
# add coeficient term
Y = cdf['cases'].values[F:(len(cdf['cases'])-N+F)]

import cvxpy as cp

# least squares problem
x = cp.Variable(N)
cost = cp.norm(X @ x - Y,2)
constraints = [x >= 0]
prob = cp.Problem(cp.Minimize(cost),constraints)
prob.solve(verbose=True)#,solver=cp.CLARABEL)

Y2 = cdf['positivity'].values[0:(len(cdf['positivity'])-N)]

# switch to logistic function
beta = cp.Variable(N,nonneg=True)#+1)
log_likelihood = cp.sum(
    cp.multiply(Y2, X @ beta) - cp.logistic(X @ beta)
)
# constraints = [beta >= 0.001, beta<=10]
prob2 = cp.Problem(cp.Maximize(log_likelihood))#,constraints)
prob2.solve(verbose=True)#,solver=cp.CLARABEL)


fig,ax = plt.subplots()
ax.plot(range(-F,len(x.value)-F),x.value) # just plotting the curve learned via the case counts for now. 
ax.set_xlabel('Date relative to detection')
ax.set_ylabel('Mean viral load copies/L')
```

```python
from scipy.linalg import toeplitz

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

# K is a Toeplitz matrix representing convolution with S
first_col = np.r_[S[0], np.zeros(len(VL)-1)]
first_row = np.r_[S, np.zeros(len(VL) - len(S))]
K = toeplitz(first_col, first_row)

# solve for C in V = S * C
lambda_reg = 0.01
I = np.eye(K.shape[1])
C = np.linalg.solve(K.T @ K + lambda_reg * I, K.T @ VL)

# align time index for plotting
C_dates = cdf.index[F:(len(cdf.index)-N+F)]


plt.figure(figsize=(10, 5))
plt.plot(C_dates, C, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y, label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cases')
```

```python
# 1. Inverting the shedding kernel, then just using regular convolution
from numpy.fft import fft, ifft

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

def inverse(VL, S, lambda_reg=0.01):
    length = len(VL) - len(S)

    # pad kernel to match length
    S_pad = np.pad(S, (0, length), mode='constant')

    # fast fourier transformation (FFT)
    VL_fft = fft(VL)
    S_fft = fft(S_pad)

    # C = ( H^T H + lambda^2 I )^-1 H^T y  =  H^T y / H^T H + lambda^2 I
    H = S_fft # shedding kernel (Fourier transformed)
    H_T = np.conj(H) # conjugate of the kernel acts like the transpose
    y = VL_fft # observed viral load (Fourier transformed)
    C = (H_T * y) / (H_T * H + lambda_reg)

    C = np.real(ifft(C))
    C[C < 0] = 0
    return C

C = inverse(VL, S)

C_dates_inv = C_dates[:len(C)]

plt.figure(figsize=(10, 5))
plt.plot(C_dates_inv, C, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y[:len(C)], label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cases')
```

```python
# 2. Performing the linear solve with a nonnegativity constraint

import cvxpy as cp

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

# K is a Toeplitz matrix representing convolution with S
first_col = np.r_[S[0], np.zeros(len(VL)-1)]
first_row = np.r_[S, np.zeros(len(VL) - len(S))]
K = toeplitz(first_col, first_row)

# solve
C = cp.Variable(K.shape[1])
cost = cp.norm(K @ C - VL, 2)
constraints = [C >= 0]

smoothness = cp.norm(cp.diff(C), 1)
prob = cp.Problem(cp.Minimize(cost + 0.1 * smoothness), constraints)
# prob = cp.Problem(cp.Minimize(cost), constraints)
prob.solve(verbose=True)

C_nnls = C.value
C_dates_nnls = C_dates[:len(C_nnls)]


plt.figure(figsize=(10, 5))
plt.plot(C_dates_nnls, C_nnls, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y[:len(C_nnls)], label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Cases")
```
