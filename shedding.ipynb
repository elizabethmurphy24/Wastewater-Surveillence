```python
from scipy.linalg import toeplitz

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

# K is a Toeplitz matrix representing convolution with S
first_col = np.r_[S[0], np.zeros(len(VL)-1)]
first_row = np.r_[S, np.zeros(len(VL) - len(S))]
K = toeplitz(first_col, first_row)

# solve for C in V = S * C
lambda_reg = 0.01
I = np.eye(K.shape[1])
C = np.linalg.solve(K.T @ K + lambda_reg * I, K.T @ VL)

# align time index for plotting
C_dates = cdf.index[F:(len(cdf.index)-N+F)]


plt.figure(figsize=(10, 5))
plt.plot(C_dates, C, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y, label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cases')
```

```python
# 1. Inverting the shedding kernel, then just using regular convolution
from numpy.fft import fft, ifft

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

def inverse(VL, S, lambda_reg=0.01):
    length = len(VL) - len(S)

    # pad kernel to match length
    S_pad = np.pad(S, (0, length), mode='constant')

    # fast fourier transformation (FFT)
    VL_fft = fft(VL)
    S_fft = fft(S_pad)

    # C = ( H^T H + lambda^2 I )^-1 H^T y  =  H^T y / H^T H + lambda^2 I
    H = S_fft # shedding kernel (Fourier transformed)
    H_T = np.conj(H) # conjugate of the kernel acts like the transpose
    y = VL_fft # observed viral load (Fourier transformed)
    C = (H_T * y) / (H_T * H)

    C = np.real(ifft(C))
    C[C < 0] = 0
    return C

C = inverse(VL, S)

C_dates_inv = C_dates[:len(C)]


plt.figure(figsize=(10, 5))
plt.plot(C_dates_inv, C, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y[:len(C)], label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel('Date')
plt.ylabel('Cases')
```

```python
# 2. Performing the linear solve with a nonnegativity constraint

import cvxpy as cp

# previously estimated kernel
S = x.value 

# viral load
VL = cdf['Mean viral gene copies/L'].values
VL = VL[F:(len(VL)-N+F)]

# K is a Toeplitz matrix representing convolution with S
first_col = np.r_[S[0], np.zeros(len(VL)-1)]
first_row = np.r_[S, np.zeros(len(VL) - len(S))]
K = toeplitz(first_col, first_row)

# solve
C = cp.Variable(K.shape[1])
cost = cp.norm(K @ C - VL, 2)
constraints = [C >= 0]

smoothness = cp.norm(cp.diff(C), 1)
prob = cp.Problem(cp.Minimize(cost + 0.1 * smoothness), constraints)
# prob = cp.Problem(cp.Minimize(cost), constraints)
prob.solve(verbose=True)

C_nnls = C.value
C_dates_nnls = C_dates[:len(C_nnls)]


plt.figure(figsize=(10, 5))
plt.plot(C_dates_nnls, C_nnls, label='Inferred Clinical Cases from Wastewater', linewidth=2)
plt.plot(C_dates, Y[:len(C_nnls)], label='Reported Clinical Cases', linewidth=2)
plt.legend()
plt.xlabel("Date")
plt.ylabel("Cases")
```

```python
# Inverse shedding kernel
from numpy.fft import fft, ifft

def inverse_kernel(S, output_len=len(S)):

    S_fft = fft(S)

    H = S_fft # shedding kernel (Fourier transformed)
    H_T = np.conj(H) # conjugate of the kernel acts like the transpose

    H_inv = H / (H_T * H)

    s_inv = np.real(ifft(H_inv))
    return s_inv

S = x.value
s_inv = inverse_kernel(S)

plt.figure(figsize=(8, 4))
plt.plot(s_inv)
plt.xlabel('Days')
plt.legend()
plt.tight_layout()
```

```python
# Shedding Kernel Might Change Over Time
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

N = 30
F = 5
# 120 day chunks, start every 60 days
window_size = 120
step_size = 60

window_starts = pd.date_range(start='2020-06-01', end='2022-07-01', freq=f'{step_size}D')
kernels = []
labels = []

for start in window_starts:
    end = start + pd.Timedelta(days=window_size)
    sub = cdf[(cdf.index >= start) & (cdf.index <= end)]
    print(f"{start.date()} to {end.date()} - data points: {len(sub)}")
    
    # Any window with fewer than 45 daily data points in cdf will be skipped
    if len(sub) < N + F + 10:
        continue

    VL = sub['Mean viral gene copies/L'].values
    cases = sub['cases'].values

    X = np.array([VL[(j - F):(j + N - F)] for j in range(F, len(VL) - N + F)])
    X = X / np.mean(VL)
    Y = cases[F:(len(cases) - N + F)]

    x = cp.Variable(N)
    cost = cp.norm(X @ x - Y,2)
    constraints = [x >= 0]
    prob = cp.Problem(cp.Minimize(cost),constraints)
    prob.solve()

    kernels.append(x.value)
    labels.append(f"{start.date()} to {end.date()}")


plt.figure(figsize=(12, 6))
for x, label in zip(kernels, labels):
    plt.plot(range(-F, len(x) - F), x, label=label)

plt.xlabel('Date relative to detection')
plt.ylabel('Mean viral load copies/L')
plt.legend()
```
